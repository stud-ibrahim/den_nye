# Weather Service - Rain Cycle Broadcaster

This notebook broadcasts a continuous weather cycle via MQTT.

**Weather Cycle:**
- 10 seconds of sunny weather (no rain)
- 5 seconds of rain
- Repeat continuously

**Usage:**
1. Run all cells to start the weather service
2. Launch person_walker.ipynb notebooks - they will react to weather changes
3. Launch map_viewer.ipynb to see the visual effects
4. Run the stop cell when finished

```python
import asyncio
import json
import time

from simulated_city.config import load_config
from simulated_city.mqtt import MqttConnector, MqttPublisher
```

```python
# Weather cycle timing (in seconds)
SUNNY_DURATION = 10.0  # Seconds of sunny weather
RAIN_DURATION = 5.0     # Seconds of rain

# MQTT topic for weather updates
WEATHER_TOPIC = "weather/rain"
```

```python
# Connect to MQTT broker from config
cfg = load_config()
connector = MqttConnector(cfg.mqtt, client_id_suffix="weather-service")
connector.connect()
if not connector.wait_for_connection(timeout=10.0):
    raise RuntimeError("Failed to connect to MQTT broker")

publisher = MqttPublisher(connector)
print("✓ Connected to MQTT broker")
print(f"  Broker: {cfg.mqtt.host}:{cfg.mqtt.port}")
print(f"  Publishing to: {WEATHER_TOPIC}")
```

```python
async def weather_cycle():
    """
    Broadcast weather cycle: 10s sunny → 5s rain → repeat.
    
    Publishes to weather/rain topic with format:
    {"raining": bool, "timestamp": float}
    
    Uses retain=True so late-joining subscribers get current weather.
    """
    cycle_count = 0
    
    print("\n🚀 Weather service started!")
    print(f"Cycle: {SUNNY_DURATION}s sunny → {RAIN_DURATION}s rain\n")
    
    while True:
        cycle_count += 1
        
        # SUNNY PERIOD (10 seconds)
        print(f"☀️  Cycle {cycle_count}: SUNNY (starts at {time.strftime('%H:%M:%S')})")
        message = {
            "raining": False,
            "timestamp": time.time()
        }
        publisher.publish_json(
            WEATHER_TOPIC,
            json.dumps(message),
            qos=0,
            retain=True  # Late-joiners get current weather state
        )
        
        await asyncio.sleep(SUNNY_DURATION)
        
        # RAIN PERIOD (5 seconds)
        print(f"☔ Cycle {cycle_count}: RAIN STARTS (at {time.strftime('%H:%M:%S')})")
        message = {
            "raining": True,
            "timestamp": time.time()
        }
        publisher.publish_json(
            WEATHER_TOPIC,
            json.dumps(message),
            qos=0,
            retain=True
        )
        
        await asyncio.sleep(RAIN_DURATION)
        print(f"   Rain ended (after {RAIN_DURATION}s)\n")
```

```python
# Start the weather cycle
weather_task = asyncio.create_task(weather_cycle())
print("✓ Weather service is running!")
print("Run the next cell to stop.")
```

```python
# Stop the weather service
weather_task.cancel()
connector.disconnect()
print("✓ Weather service stopped.")
```
